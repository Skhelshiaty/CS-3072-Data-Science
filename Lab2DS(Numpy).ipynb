### Exercise 1.4: Array Indexing and Slicing
Practice accessing and modifying array elements.

```python
# Create a sample grade matrix (5 students, 4 subjects)
grade_matrix = np.array([
    [85, 90, 88, 92],  # Student 1
    [78, 82, 80, 85],  # Student 2
    [92, 95, 91, 94],  # Student 3
    [88, 87, 89, 86],  # Student 4
    [91, 93, 90, 88]   # Student 5
])

print("Grade Matrix:")
print(grade_matrix)

# TODO: Get the grades for student 3 (index 2)
student3_grades = grade_matrix[2]
print("\nStudent 3 grades:", student3_grades)

# TODO: Get the grades for subject 2 (index 1) for all students
subject2_grades = grade_matrix[:, 1]
print("Subject 2 grades for all students:", subject2_grades)

# TODO: Get grades for the first 3 students in the first 2 subjects
subset_grades = grade_matrix[:3, :2]
print("\nSubset grades (first 3 students, first 2 subjects):")
print(subset_grades)

# QUIZ EXPLANATION:
# grade_matrix[row, column]
# : means "all"
# grade_matrix[:, 1] -> all rows, column 1 (all students for subject 2)
# grade_matrix[2, :] -> row 2, all columns (all grades for student 3)
```

### Exercise 2.2: Array Statistics and Mathematical Functions
Calculate various statistics from the grade data.

```python
# TODO: Calculate overall statistics for the grade matrix
overall_mean = np.mean(grade_matrix)
overall_std = np.std(grade_matrix)
overall_max = np.max(grade_matrix)
overall_min = np.min(grade_matrix)

print(f"Overall Statistics:")
print(f"Mean: {overall_mean:.2f}")
print(f"Standard Deviation: {overall_std:.2f}")
print(f"Maximum: {overall_max}")
print(f"Minimum: {overall_min}")

# TODO: Calculate mean grade for each student (across subjects)
student_means = np.mean(grade_matrix, axis=1)
print(f"\nMean grades per student: {student_means}")

# TODO: Calculate mean grade for each subject (across students)
subject_means = np.mean(grade_matrix, axis=0)
print(f"Mean grades per subject: {subject_means}")

# QUIZ EXPLANATION:
# axis tells NumPy the direction of calculation
# axis=0 -> operate down columns (per subject)
# axis=1 -> operate across rows (per student)
```

## Section 6: Data Cleaning and Manipulation

### Exercise 6.1: Introducing Missing Values and Cleaning
Add some missing values to practice data cleaning techniques.

```python
# Create a copy of the dataset and introduce some missing values
df_with_missing = df_large.copy()

# Randomly set some values to NaN
np.random.seed(42)
missing_indices = np.random.choice(df_with_missing.index, size=20, replace=False)
df_with_missing.loc[missing_indices[:10], 'Math'] = np.nan
df_with_missing.loc[missing_indices[10:15], 'StudyHours'] = np.nan
df_with_missing.loc[missing_indices[15:], 'Attendance'] = np.nan

# TODO: Check for missing values
print("Missing values per column:")
print(df_with_missing.isnull().sum())

# TODO: Display rows with missing values
print("\nRows with missing values:")
print(df_with_missing[df_with_missing.isnull().any(axis=1)])
```

### Exercise 6.2: Handling Missing Values
Practice different strategies for dealing with missing data.

```python
# TODO: Fill missing Math scores with the mean
math_mean = df_with_missing['Math'].mean()
df_with_missing['Math'].fillna(math_mean, inplace=True)

# TODO: Fill missing StudyHours with the median
study_hours_median = df_with_missing['StudyHours'].median()
df_with_missing['StudyHours'].fillna(study_hours_median, inplace=True)

# TODO: Fill missing Attendance with forward fill method
df_with_missing['Attendance'].fillna(method='ffill', inplace=True)

# TODO: Verify that missing values have been handled
print("Missing values after cleaning:")
print(df_with_missing.isnull().sum())

# QUIZ EXPLANATION:
# Mean is used for normal numerical data
# Median is used when data may have outliers
# Mode is used when data is catagorical
# fillna() replaces missing (NaN) values with chosen statistic
```

### Exercise 6.3: Data Filtering and Selection
Practice filtering data based on conditions.

```python
# TODO: Filter students with Math scores above 85
high_math_students = df_with_missing[df_with_missing['Math'] > 85]
print(f"Students with Math > 85: {len(high_math_students)}")
print(high_math_students[['Name', 'Math']].head())

# TODO: Filter female students with high attendance (>0.9)
high_attendance_females = df_with_missing[
    (df_with_missing['Gender'] == 'Female') & (df_with_missing['Attendance'] > 0.9)
]
print(f"\nFemale students with attendance > 0.9: {len(high_attendance_females)}")

# TODO: Filter students who study more than 25 hours and have parent with Bachelor+ education
studious_educated_parents = df_with_missing[
    (df_with_missing['StudyHours'] > 25) &
    (df_with_missing['ParentEducation'].isin(['Bachelor', 'Master', 'PhD']))
]
print(f"\nStudious students with educated parents: {len(studious_educated_parents)}")
```

## Section 7: Data Analysis and Statistics

### Exercise 7.1: Descriptive Statistics
Calculate various statistics for the student data.

```python
# TODO: Calculate overall GPA for each student (average of Math, Science, English, History)
df_with_missing['GPA'] = df_with_missing[['Math', 'Science', 'English', 'History']].mean(axis=1)

print("GPA Statistics:")
print(df_with_missing['GPA'].describe())

# TODO: Find the top 10 students by GPA
top_students = df_with_missing.sort_values(by='GPA', ascending=False).head(10)
print("\nTop 10 Students by GPA:")
print(top_students[['Name', 'GPA']])
```

### Exercise 7.3: Correlation Analysis
Examine relationships between different variables.

```python
# TODO: Calculate correlation matrix for numerical columns
numerical_cols = ['Age', 'Math', 'Science', 'English', 'History', 'StudyHours', 'Attendance', 'GPA']
correlation_matrix = df_with_missing[numerical_cols].corr()

print("Correlation Matrix:")
print(correlation_matrix)

# TODO: Find the strongest correlations with GPA
gpa_correlations = correlation_matrix['GPA']
print("\nCorrelations with GPA (sorted):")
print(gpa_correlations.sort_values(ascending=False))

# QUIZ TIP:
# Correlation shows relationship strength between variables
# Values close to 1 or -1 mean strong relationship
# GPA is expected to correlate with subject scores
```

### Exercise 7.4: Function Application
Apply custom functions to the data.

```python
# TODO: Create a function to categorize GPA into letter grades
def gpa_to_letter_grade(gpa):
    # A: >= 90, B: >= 80, C: >= 70, D: >= 60, F: < 60
    if gpa >= 90:
        return 'A'
    elif gpa >= 80:
        return 'B'
    elif gpa >= 70:
        return 'C'
    elif gpa >= 60:
        return 'D'
    else:
        return 'F'

# TODO: Apply the function to create a new column
df_with_missing['LetterGrade'] = df_with_missing['GPA'].apply(gpa_to_letter_grade)

# TODO: Count the distribution of letter grades
grade_distribution = df_with_missing['LetterGrade'].value_counts()
print("Letter Grade Distribution:")
print(grade_distribution)

# TODO: Create a function to calculate the range (max - min) for each student's scores
def score_range(row):
    scores = [row['Math'], row['Science'], row['English'], row['History']]
    return max(scores) - min(scores)

# TODO: Apply the function across rows
df_with_missing['ScoreRange'] = df_with_missing.apply(score_range, axis=1)

print("\nScore Range Statistics:")
print(df_with_missing['ScoreRange'].describe())
```
